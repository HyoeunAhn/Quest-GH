```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('MovieLense.csv')
df
```

### User based CF

```python
#결측치를 각 유저가 준 평균 점수로 채우기
df = df.apply(lambda x: x.fillna(x.mean()),axis=1)
df
```

```python
from sklearn.metrics.pairwise import cosine_similarity  

user_cosine_sim = pd.DataFrame(cosine_similarity(df))
user_cosine_sim
```

```python
user_mean = df.mean(axis=1)
item_mean = df.mean(axis=0)
user_cosine_sum = user_cosine_sim.sum()
```

```python
def user_score_function(user, item):
    numerator = 0
    
    for v in range(943):
        numerator += (df[item][v]-user_mean[v])*(user_cosine_sim[user][v])
        score = user_mean[user] + numerator/user_cosine_sum[user]
    
    return score
```

```python
user_score_function(316, 'Toy Story (1995)')
```

```python
Movie = df.columns
Movie
```

```python
#user316의 user_score_function 결과
import time
t0 = time.time()
user_316 = []

for item in Movie:
    user_316.append(user_score_function(316, item))
        
user_316 = pd.DataFrame(user_316, index = Movie)
print("소요시간: {:.3f} sec".format(time.time()-t0))
```

시간이 오래 걸리는 것 같아서 한번 측정해봤는데 정말 오래걸리네요ㅠㅠ

### Item based CF

```python
df2 = pd.read_csv('MovieLense.csv')
df2 = df2.T

#drop을 하면 차원이 너무 작아지기 때문에 0.5로 채운다.
df2=df2.fillna(0.5)
```

```python
item_cosine_sim = pd.DataFrame(cosine_similarity(df2))
item_cosine_sim
```

```python
(item_cosine_sim < 0).values.any()
#코사인 유사도의 모든 값이 양수이기 때문에 score function의 분자를 계산할 때 별다른 변환을 하지 않아도 된다.
```

```python
item_mean2 = df2.mean(axis=1)
user_mean2 = df2.mean(axis=0)
item_cosine_sum = item_cosine_sim.sum()
```

```python
movie_list = Movie.tolist()
movie_list
```

```python
def item_score_function(item, user):
    numerator = 0
    for j in range(1664):
        numerator += (item_cosine_sim[j][movie_list.index(item)])*(df2[user][j]-item_mean2[j])
        score = item_mean2[item] + numerator/item_cosine_sum[movie_list.index(item)]
    
    return score
```

```python
item_score_function('Toy Story (1995)', 316)
```

```python
item_score_function('GoldenEye (1995)', 316)
```

```python
#user316의 item_score_function 결과
t0 = time.time()
item_316 = []

for item in Movie:
    item_316.append(item_score_function(item, 316))
        
item_316 = pd.DataFrame(item_316, index=Movie)
print("소요시간: {:.3f} sec".format(time.time()-t0))
```
